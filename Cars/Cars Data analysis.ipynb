## Data cleaning

#### 1 - Find all the Null values in the dataset. if there are any null values in any column, fil it with the mean of that column

```python
data.isnull().sum()
```

```python
# Replacing null values with the mean of that column
data.fillna(data.mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Find columns that has objcts
object_col = data.select_dtypes(include=['object']).columns
object_col
```

```python
# drop the null values in those columns
data.dropna(subset=object_col, inplace=True)
```

```python
data.isnull().sum()
```
